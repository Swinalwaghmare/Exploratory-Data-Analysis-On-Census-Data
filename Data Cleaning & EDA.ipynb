The census dataset appears to be the U.S. Census Bureau's Current Population Survey (CPS) dataset. This dataset includes information on various demographic and socio-economic characteristics of individuals in the United States.

Here is a brief description of each of the features you mentioned:

1. Age: The age of the individual.
2. Workclass: The type of employer or employment status of the individual, such as "Private", "Self-emp-not-inc", "Local-gov", etc.
3. Fnlwgt: A weighting factor used to adjust for the sampling design of the CPS survey.
4. Education: The highest level of education attained by the individual.
5. Education-num: A numerical representation of the education level of the individual, with higher numbers indicating more education.
6. Marital-status: The current marital status of the individual, such as "Married-civ-spouse", "Divorced", "Never-married", etc.
7. Occupation: The type of job or occupation of the individual, such as "Exec-managerial", "Craft-repair", "Sales", etc.
8. Relationship: The individual's relationship to the household head, such as "Husband", "Wife", "Own-child", "Unmarried", etc.
9. Race: The individual's race or ethnic background, such as "White", "Black", "Asian-Pac-Islander", "Other", etc.
10. Sex: The individual's gender.
11. Capital-gain: The amount of capital gains reported by the individual in the previous year's tax return.
12. Capital-loss: The amount of capital losses reported by the individual in the previous year's tax return.
13. Hours-per-week: The number of hours the individual works per week.\
This dataset is often used for tasks such as predicting an individual's income level based on their demographic and socio-economic characteristics, or analyzing trends in employment and education.

## Numerical Features

```python
# Numerical features
numerical_feature = [feature for feature in df.columns if df[feature].dtypes != "O"]
```

```python
# Distribution plot of Numerical features
plt.figure(figsize=(14,7))
plt.suptitle('Numerical features',fontsize=20, fontweight = 'bold', alpha=0.8,y=1)

for i in range(0, len(numerical_feature)):
    plt.subplot(3,2,i+1)
    sns.kdeplot(x=df[numerical_feature[i]],shade=True,color='lime')
    plt.xlabel(numerical_feature[i])
    plt.tight_layout()
```

### Age

```python
# plotting the age graph w.r.t class
plt.figure(figsize=(10,5))
sns.kdeplot(data=df,x='Age',hue="class",fill=True)
plt.title("Age Distribution")
plt.show()
```

- The age distribution is positively skeweed.
- As we can see from the first graph, the age distribution for the <=50k income class is heavily right-skewed. There could be several reasons for this, such as a lack of a strong educational background or the fact that they may have just gotten out of college and had their first job.
- The second graph,the distribtion for the >50k income class is following kind of normal distribution.

## Categorical Features 

```python
# categorical features
categorical_feature = [feature for feature in df.columns if df[feature].dtypes == "O"]
```

```python
# catrgorical columns values are having spacing issues removing with user defined function
def strip_columns(data,columns):
    for col in columns:
        data.loc[:,col] = data.loc[:,col].str.strip()
```

```python
# calling function on categorical features
strip_columns(df,categorical_feature)
```

```python
# unique value counts for each categorical variable 
for col in categorical_feature:
    print(f'{col} :\n{df[col].value_counts()}')
    print("\n")
```

### Sex

```python
# Plotting pie plot on sex column
df['sex'].value_counts().plot.pie(y=df['sex'],figsize=(10,8),autopct='%1.1f%%')
plt.show()
```

- In theis pie plot we can see that the imbalence between sex column only 1/3 of the total individuals are female.
- Hence we can make the analysis on both the gender respectively.

```python
# Plotting pieplot on sex column on male and female categorys w.r.t class
plt.subplot(1,2,1)
df[df['sex']=='Male']['class'].value_counts().plot.pie(y=df[df['sex']=='Male']['class'],
                                    startangle=0,figsize=(10,8),
                                    autopct='%1.1f%%',textprops={'fontsize': 14})
plt.title("Male Income Distribution")
plt.subplot(1,2,2)
df[df['sex']=='Female']['class'].value_counts().plot.pie(y=df[df['sex']=='Female']['class'],
                                    startangle=0,figsize=(10,8),
                                    autopct='%1.1f%%',textprops={'fontsize': 14})
plt.title("Female Income Distribution")
plt.show()
```

- In the Male Group around 30% people are earning >50K.
- In Female Group around 11% people are earning >50K.
- This can mean that there will exist some gap between the income of male and female group.
